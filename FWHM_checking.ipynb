```
import glob
from google.colab import drive
from google.colab import files
drive.mount('/content/drive')
import numpy as np
import matplotlib.pyplot as plt
```

```
# Extraction of the CSV file into array form

float_data = []
image_no = 0
image_no_list =[]
p = open('/content/drive/MyDrive/photometry_fwhm_v432per_10.csv', mode = 'r')
file_to_split = p.read()
for i, row in enumerate(file_to_split.split('\n')):
  if i==0 or not row:
    continue
  splitted_row = row.split(',')
  float_row = [float(p) for p in splitted_row[1:]]
  float_data.append(float_row)
print(float_data)
fwhm_data = np.array(float_data)
```

```
#Plot of the FWHM data- Image number

fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.plot(range(len(fwhm_data[:,0])), fwhm_data[:,0], marker='o', linestyle='dashed', color='blue', label='vertical fwhm')
ax.plot(range(len(fwhm_data[:,2])), fwhm_data[:,2], marker='o', linestyle='dashed', color='red', label='horizontal fwhm')
leg = ax.legend();
plt.xlabel('Image No')
plt.ylabel('FWHM')
```

```
# Plot of the x0 data-Image number

fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.plot(range(len(fwhm_data[:,1])),fwhm_data[:,1], marker='o', linestyle='dashed', color='blue', label='vertical x0')
ax.plot(range(len(fwhm_data[:,3])),fwhm_data[:,3], marker='o', linestyle='dashed', color='red', label='horizontal x0')
leg = ax.legend();
plt.xlabel('Image No')
plt.ylabel('x0')
```
